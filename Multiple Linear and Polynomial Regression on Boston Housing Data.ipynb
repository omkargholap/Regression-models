```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read dataset
df=pd.read_csv("F:\\symb. ML\\data files\\housing.csv")
df.head()
```

```python
df.info()
```

```python
# check null values
df.isna().sum()
```

```python
# bivariate analysis
sns.pairplot(x_vars=['0','RM', 'LSTAT', 'PTRATIO'],y_vars=["MEDV"],data=df)
```

```python
# correlation plot
sns.heatmap(df.corr(),annot=True,linewidth=1)
```

```python
# independent and dependent variable
x=df.drop(columns=["MEDV","PTRATIO"])
y=df["MEDV"]
```

```python
# split data 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# fit regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
# predict using test data
ypred=model.predict(xtest)
```

```python
# Evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(ytest,ypred)
print("MSE:",mse)
print("RMSE:",np.sqrt(mse))
mae=mean_absolute_error(ytest,ypred)
print("MAE:",mae)
score=r2_score(ytest,ypred)
print("R2 score:",score)
```

### Standard Scaler for Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(xtrain)
sc_xtrain=sc.transform(xtrain)
sc_xtest=sc.transform(xtest)
```

```python
model=LinearRegression()
model.fit(sc_xtrain,ytrain)
```

```python
sc_ypred=model.predict(sc_xtest)
```

```python
mse=mean_squared_error(ytest,sc_ypred)
print("MSE:",mse)
print("RMSE:",np.sqrt(mse))
mae=mean_absolute_error(ytest,sc_ypred)
print("MAE:",mae)
score=r2_score(ytest,sc_ypred)
print("R-squared :",score)
```

### Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
mm.fit(xtrain)
mm_xtrain=mm.transform(xtrain)
mm_xtest=mm.transform(xtest)
```

```python
model=LinearRegression()
model.fit(mm_xtrain,ytrain)
```

```python
mm_ypred=model.predict(mm_xtest)
```

```python
mse=mean_squared_error(ytest,mm_ypred)
print("MSE:",mse)
print("RMSE:",np.sqrt(mse))
mae=mean_absolute_error(ytest,mm_ypred)
print("MAE:",mae)
score=r2_score(ytest,mm_ypred)
print("R-squared :",score)
```

### Scaling of data does not impact the results. So, linear regression is not working for this data.

## Polynomial Regression

```python
# fit model using polynomial features
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=5) 
poly_features=pf.fit_transform(xtrain)
poly_features
```

```python
model=LinearRegression()
model.fit(poly_features,ytrain)
```

```python
pf_ypred=model.predict(pf.fit_transform(xtest))
```

```python
mse=mean_squared_error(ytest,pf_ypred)
print("MSE:",mse)
print("RMSE:",np.sqrt(mse))
mae=mean_absolute_error(ytest,pf_ypred)
print("MAE:",mae)
score=r2_score(ytest,pf_ypred)
print("R-squared :",score)
```

### Polynomial regression model is giving better performance than Linear regression model.
